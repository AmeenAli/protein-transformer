```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from  matplotlib.ticker import FuncFormatter
import seaborn as sns
sns.set()
from glob import glob
%matplotlib inline
mpl.rcParams['figure.dpi'] = 200
mpl.rcParams['figure.figsize'] = [8, 6]
```

```python
def plot_batch_loss(df, mname=""):
    x = range(len(df[df.is_val == False]["loss"]))
    sns.lineplot(x=x, y=df[df.is_val == False]["loss"], label="loss")
#     sns.lineplot(x=x, y=df[df.is_val == False]["nloss"], label="norm-loss")
    plt.xlabel("Iteration")
    plt.ylabel("DRMSD Loss")
    if mname != "":
        plt.title(mname)
    plt.legend()
    
def plot_epoch_loss(df, mname="", train_only=False, line_label="train"):
    val_epoch_loss = df[df.is_val == True]["loss"]
    train_epoch_loss = df[(df.is_end_of_epoch == True) & (df.is_val == False)]["loss"]
    if not train_only: plt.plot(range(len(val_epoch_loss)), val_epoch_loss, label="val")
    plt.plot(range(len(val_epoch_loss)), train_epoch_loss, label=line_label)
    plt.xlabel("Epoch")
    plt.ylabel("DRMSD Loss")
    if mname != "":
        plt.title(mname)
    
```

```python
train_logs = glob("../logs/0422*.train")
train_log_dfs = []
for tl in train_logs:
    df = pd.read_csv(tl)
    mname = tl.split("/")[-1][:-6]
    plot_batch_loss(df, "Model "+ mname)
    plt.figure()
#     plot_epoch_loss(df, "Model "+ mname)
#     plt.legend()
    plt.figure()
```

```python
train_logs = glob("../logs/0422*.train")
train_log_dfs = []
for tl in sorted(train_logs):
    df = pd.read_csv(tl)
    mname = tl.split("/")[-1][:-6]
    train_epoch_loss = df[(df.is_end_of_epoch == True) & (df.is_val == False)]["loss"]
    plt.plot(range(len(train_epoch_loss)), train_epoch_loss, label=mname)
plt.xlabel("Epoch")
plt.ylabel("DRMSD Loss")
plt.title("Loss Comparison")
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
# plt.show()
plt.savefig("../analysis/190423/all_losses.png")
```
